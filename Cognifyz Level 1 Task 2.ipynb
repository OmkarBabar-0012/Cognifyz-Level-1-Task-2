# Task-2: Descriptive Analysis

```python
import pandas as pd

# Load the dataset
file_path = r'C:\Users\91702\Dataset .csv'
data = pd.read_csv(file_path)

# Calculate basic statistical measures for numerical columns
numerical_columns = ['Average Cost for two', 'Longitude', 'Latitude', 'Price range', 'Aggregate rating', 'Votes']
statistics = data[numerical_columns].describe().T
statistics['median'] = data[numerical_columns].median()
print("Basic Statistical Measures:")
print(statistics)

# Distribution of Country Code
country_code_distribution = data['Country Code'].value_counts()
print("\nDistribution of Country Code:")
print(country_code_distribution)

# Distribution of City (Top 10)
city_distribution = data['City'].value_counts().head(10)
print("\nDistribution of City (Top 10):")
print(city_distribution)

# Distribution of Cuisines (Top 10)
cuisines_distribution = data['Cuisines'].value_counts().head(10)
print("\nDistribution of Cuisines (Top 10):")
print(cuisines_distribution)

# Identify top cuisines and cities with the highest number of restaurants
top_cuisines = data['Cuisines'].value_counts().head(10)
top_cities = data['City'].value_counts().head(10)
print("\nTop Cuisines:")
print(top_cuisines)
print("\nTop Cities with Highest Number of Restaurants:")
print(top_cities)
```
